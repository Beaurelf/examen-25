```python
produts = {pid : [] for pid in products_df['ProductID'].unique()}
for pid in produts:
    orderIds = df = orders_df[orders_df['ProductID'] == pid]['OrderID'].unique().tolist()
    pids = []
    for orderId in orderIds:
        pids = orders_df[orders_df['OrderID'] == orderId]['ProductID'].unique().tolist()
        pids.remove(pid)
        s = ', '.join(pids)
        produts[pid].append(s)
```

```python
f = []
for pid in produts:
    for p in produts[pid]:
        f.append([pid, p])
Q1_df = pd.DataFrame(f, columns=['ProductID', 'FrequentlyBoughtTogether'])
Q1_df
```
